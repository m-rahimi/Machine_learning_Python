```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Data/score3.csv")
```

```python
df["3Months"] = df["3Months"].values * 1000
df["CV10"] = df["CV10"].values * 1000
df["LB"] = df["LB"].values * 1000
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
color = sns.color_palette()

%matplotlib inline
```

```python
df = df.drop(0)
```

```python
sns.set(font_scale=1.5)
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots(figsize=(8,6))
points = ax.scatter(df["3Months"].values, df["CV10"].values, c=df["LB"].values, s=50, cmap=cmap)
f.colorbar(points)

plt.xlabel('3Months', fontsize=18)
plt.ylabel('CV (10 folds)', fontsize=18)
```

```python
x, y, z = np.random.rand(3, 100)
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots()
points = ax.scatter(x, y, c=z, s=50, cmap=cmap)
f.colorbar(points)
```
